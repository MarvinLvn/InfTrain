```python
import json
import os
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from glob import glob
#import seaborn as sns
```

```python
#better and bigger plots inline
plt.rcParams['figure.figsize'] = [12, 8]
plt.rcParams['figure.dpi'] = 200 # 200 e.g. is really fine, but slower
```

```python
HOME="/linkhome/rech/genini01/ucv88ce/"
eval_path=os.path.join(HOME,"projects/MultilingualCPC/experiments/eval/inftrain" )
```

```python
def retrieve_scores(n_fam=32, size="200h", eval_path='/linkhome/rech/genini01/ucv88ce/projects/MultilingualCPC/experiments/eval/inftrain'):
    fam_dict={}

    #columns = fam_id, size, eval_lang, en_model, fr_model, bil_model, same, diff (same and diff are just the scores from tje model corresponding to th train and eval)
    #do average of within and across

    for fam in ["{}".format(str(x).zfill(2)) for x in range(n_fam-1)] :
        for eval_lang in ['english','french']:
            k = fam+"_"+eval_lang

            values = [fam, size, eval_lang]

            for model in ["EN","FR","EN+FR"]:
                path = os.path.join(eval_path,model,size,fam,"cpc_small/ABX","*",eval_lang, "ABX_scores.json")
                json_f = glob(path)
                if not json_f :
                    score = np.nan
                elif len(json_f) > 1:
                    raise ValueError("multiple checkpoint evaluated in {}".format(os.path.join(eval_path,"EN",size,fam,"cpc_small/ABX")))
                else:
                    with open(json_f[0], "r") as f:

                        scores = json.load(f)
                    score = np.mean([scores['within'],scores['across']]) * 100
                values.append(score)
                
                if model == "EN" and eval_lang == "english" :
                    same = score
                elif model == "EN" and eval_lang == "french" :
                    different = score
                elif model == "FR" and eval_lang == "french" :
                    same = score
                elif model == "FR" and eval_lang == "english" :
                    different = score
            
            values.append(same)
            values.append(different)
            fam_dict[k] = values

    df = pd.DataFrame.from_dict(fam_dict, orient='index', columns=["fam_id", "size", "eval_lang", "en_model", "fr_model", "bil_model", 'same', 'different'])
    return df
```

```python
df_200h = retrieve_scores(size="200h")
df_100h = retrieve_scores(size="100h")
df = df_100h.append(df_200h, ignore_index=True)
```
