```python
def retrieve_abx_score(detailed_abx):
    # groupby middle-phone, speaker and context
    detailed_abx = detailed_abx.groupby(['middle-phone-a', 'middle-phone-b', 'middle-phone-x', 'context', 'speaker-a', 'speaker-b', 'speaker-x']) \
                .mean() \
                .reset_index()
    # groupby middle-phone and speaker
    detailed_abx = detailed_abx.groupby(['middle-phone-a', 'middle-phone-b', 'middle-phone-x', 'speaker-a', 'speaker-b', 'speaker-x']) \
                .mean() \
                .reset_index()
    # groupby middle-phone
    detailed_abx = detailed_abx.groupby(['middle-phone-a', 'middle-phone-b', 'middle-phone-x']) \
                .mean() \
                .reset_index()
    abx = detailed_abx['model_score'].mean()
    return abx
```

```python
phone_abx_EN = detailed_abx_within_en.groupby(['middle-phone-x']).agg(
             count=('model_score', 'count'),
             model_score=('model_score', 'mean'))

detailed_abx_within_en["contrast"] =  ['+'.join(sorted(tup)) for tup in zip(detailed_abx_within_en['middle-phone-a'], detailed_abx_within_en['middle-phone-b'])]
contrast_groups_EN = detailed_abx_within_en.groupby(['contrast']).agg(
             count=('model_score', 'count'),
             model_score=('model_score', 'mean'))
contrast_groups_EN["lang"] ="en"


phone_abx_FR = detailed_abx_within_fr.groupby(['middle-phone-x']).agg(
             count=('model_score', 'count'),
             model_score=('model_score', 'mean'))

detailed_abx_within_fr["contrast"] = ['+'.join(sorted(tup)) for tup in zip(detailed_abx_within_fr['middle-phone-a'], detailed_abx_within_fr['middle-phone-b'])]
contrast_groups_FR = detailed_abx_within_fr.groupby(['contrast']).agg(
             count=('model_score', 'count'),
             model_score=('model_score', 'mean'))
contrast_groups_FR["lang"] ="fr"
```
