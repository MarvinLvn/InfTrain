```python
# Read item file
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os

plt.rcParams['figure.figsize'] = [12, 8]
plt.rcParams['figure.dpi'] = 200 # 200 e.g. is really fine, but slower
```

```python
lang="fr" #en
ITEM_PATH = os.path.join("/gpfsscratch/rech/cfs/commun/cv21_ABX/raw_dataset",lang,lang+".stats")
df = pd.read_csv(ITEM_PATH, sep=" ")
df['spk'] = df.apply(lambda x : x['speaker'][0:5], axis=1)
nb_phones = len(np.unique(df["#phone"]))


print("%d different phones found" % nb_phones)
```

```python
print(len(np.unique(df["spk"])))
print(len(np.unique(df["speaker"])))
```

```python
df
```

```python
phone_distrib = df["#phone"].value_counts()
plt.figure(figsize=(20,10))
plt.bar(phone_distrib.index, phone_distrib)
plt.xlabel("Phone")
plt.ylabel("Count")
plt.title("Center phone distribution in triphone items for French")
plt.xticks(rotation=45,fontsize=18)
plt.show()
```

```python
spk_distrib = df["spk"].value_counts()
plt.figure(figsize=(20,10))
plt.bar(spk_distrib.index, spk_distrib)
plt.xlabel("Speaker")
plt.ylabel("Count")
plt.title("Speaker distribution for French")
plt.xticks(rotation=45,fontsize=18)
plt.show()
```

```python
df_phonedur = df.groupby(by="#phone").mean().sort_values(by="center-phone-dur", ascending=False)
df_phonedur.plot.bar(y="center-phone-dur")
print("Average general phone duration (on average of each phone duration): {} ms".format(round(df_phonedur["center-phone-dur"].mean()*1000)))
```
